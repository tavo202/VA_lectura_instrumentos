```python
image_file_name = "Imagen.png"

# --- Image Loading and Error Handling ---
image = cv2.imread(image_file_name)

# Crucial check: Verify if the image was loaded successfully
if image is None:
    print(f"Error: Could not load image from '{image_file_name}'.")
    print("Please ensure the file exists and the path is correct.")
    # You might want to exit or raise an error here if the image is critical
    exit()

# --- Image Processing ---
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convert BGR (OpenCV) to RGB (Matplotlib)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)     # Convert to grayscale

# --- Display with Matplotlib ---
print("Displaying Original RGB Image with Matplotlib...")
plt.figure(figsize=(10, 6)) # Create a figure with a specified size
plt.imshow(image_rgb)      # Show the RGB image
plt.title("Original Image") # Set plot title
plt.axis("off")            # Hide axes for a cleaner image display
plt.show()                 # Display the Matplotlib plot

```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\Gustavo\AppData\Local\Programs\Tesseract-OCR\tesseract'
extracted_text = pytesseract.image_to_string(image_rgb)
print(" Extracted Text:\n")
print(extracted_text)
```

```python
image_file_name = "temp.jpg"

# --- Image Loading and Error Handling ---
image = cv2.imread(image_file_name)

# Crucial check: Verify if the image was loaded successfully
if image is None:
    print(f"Error: Could not load image from '{image_file_name}'.")
    print("Please ensure the file exists and the path is correct.")
    # You might want to exit or raise an error here if the image is critical
    exit()

# --- Image Processing ---
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convert BGR (OpenCV) to RGB (Matplotlib)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)     # Convert to grayscale

# --- 5. Visualizar las im√°genes con Matplotlib ---
# Crea una figura con dos subplots (1 fila, 2 columnas)
plt.figure(figsize=(12, 6)) # Define el tama√±o de la figura para una mejor visualizaci√≥n

# Subplot 1: Imagen Original (RGB)
plt.subplot(1, 2, 1) # (filas, columnas, √≠ndice_del_plot)
plt.imshow(image_rgb)
plt.title("Imagen Original")
plt.axis("off") # Oculta los ejes para una vista m√°s limpia

# Subplot 2: Imagen en Escala de Grises
plt.subplot(1, 2, 2) # (filas, columnas, √≠ndice_del_plot)
plt.imshow(gray, cmap='gray') # 'cmap='gray'' es crucial para que Matplotlib la muestre correctamente en grises
plt.title("Imagen en Escala de Grises")
plt.axis("off") # Oculta los ejes

# Ajustar el dise√±o para evitar superposiciones y mostrar la figura
plt.tight_layout()
plt.show()

pytesseract.pytesseract.tesseract_cmd = r'C:\Users\Gustavo\AppData\Local\Programs\Tesseract-OCR\tesseract'
extracted_text = pytesseract.image_to_string(gray)
print(" Extracted Text:\n")
print(extracted_text)
```
