```python
import cv2
#import pytesseract
from matplotlib import pyplot as plt

#pytesseract.pytesseract.tesseract_cmd = r'C:\Users\Gustavo\AppData\Local\Programs\Tesseract-OCR\tesseract'

image_file_name = "temp.jpg"
image = cv2.imread(image_file_name)
if image is None:
    print(f"Error: no se pudo cargar la imagen '{image_file_name}'")
    exit()

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# === 1. Umbral fijo (inverso) ===
_, thresh_fixed = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)

# === 2. Umbral adaptativo ===
thresh_adaptive = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY_INV,
    11, 2
)

# Mostrar las 3 versiones
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(thresh_fixed, cmap='gray')
plt.title("Umbral fijo INV")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(thresh_adaptive, cmap='gray')
plt.title("Umbral adaptativo")
plt.axis("off")

plt.tight_layout()
plt.show()

# Puedes probar OCR en ambos para ver cuál funciona mejor
config = r'--oem 3 --psm 6 -c tessedit_char_whitelist=0123456789.-'

#text_fixed = pytesseract.image_to_string(thresh_fixed, config=config)
#text_adaptive = pytesseract.image_to_string(thresh_adaptive, config=config)

#print("Texto con umbral fijo INV:\n", text_fixed)
#print("Texto con umbral adaptativo:\n", text_adaptive)
```
